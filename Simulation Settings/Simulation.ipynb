```python
U = 80 #ft/sec
c = 1 #feet choord
visc = 1
#rho = P / (R*T)
rho = .0749
b = 2.8
Ap = c * b
```

## Values

```python


U = 80  # Free stream velocity
max_angle = 25 * mt.pi / 180  # Max angle in radians
theta = np.linspace(0, max_angle, 6)  # Angle array

# Velocity components
x_vel = np.cos(theta) * U
y_vel = np.sin(theta) * U  # Negative for downward direction

# Print each column separately
print("Angle (deg):")
for angle in theta:
    print(f"{angle * 180 / mt.pi:.6f}")

print("\nX-Velocity:")
for x in x_vel:
    print(f"{x:.6f}")

print("\nY-Velocity:")
for y in y_vel:
    print(f"{y:.6f}")




```

```python
#Results

L_D_Fluent = np.array([
[18.681, 1.3289],    
[16.129955,	1.386283],
[18.316593,	1.3184408],
[17.041377,	1.2551106],
[26.761019,	1.1112823],
[30.127032,	0.95528593]]
)

L_Fluent = L_D_Fluent[:,0] * .22448
D_Fluent = L_D_Fluent[:,1] * .22448
L = L_Fluent*np.cos(theta) - D_Fluent*np.sin(theta) 
D = L_Fluent*np.sin(theta) + D_Fluent*np.cos(theta)

C_L = 2*L / (rho * U**2 * Ap)
C_D = 2*D / (rho * U ** 2 * Ap)

print(C_L,C_D)

```

```python
fig, ax = plt.subplots()
ax.plot(C_D,C_L)
ax.set_ylabel("C_L")
ax.set_xlabel("C_D")
```
